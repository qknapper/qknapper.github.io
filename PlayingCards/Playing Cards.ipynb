```python
import pandas as pd
import datetime as dt
```

```python
collection = pd.read_excel("Collection.xlsx")
collection
```

```python
videoFormat = "\t\t\t<video class=\"playingCards\" poster=\"PlayingCards/Backs/{cardID}.jpg\" controls>" + "\n" + \
                    "\t\t\t\t<source src=\"PlayingCards/Videos/{cardID}.mp4\" type=\"video/mp4\">" + "\n" + \
                "\t\t\t</video>"

imageFormat = "\t\t\t<img class=\"playingCards\" src=\"PlayingCards/Backs/{cardID}.jpg\">"
```

```python
videoLink = "<a href=\"PlayingCards/Videos/{cardID}.mp4\" target=\"_blank\">Video <span>({cardID}.mp4)</span></a>"
videoNoLink = "<span class=\"fakeLink\">No Video <span>(for now...)</span></span>"

tableFormat = "\t\t\t<table class=\"playingCards\">" + "\n" + \
                    "\t\t\t\t<tr>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\"><span class=\"featureName\">Style: </span>{cardStyle}</td>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\"><span class=\"featureName\">Publisher: </span>{cardPublisher}</td>" + "\n" + \
                    "\t\t\t\t</tr>" + "\n" + \
                    "\t\t\t\t<tr>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\"><span class=\"featureName\">Stock: </span>{cardStock}</td>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\"><span class=\"featureName\">Color: </span>{cardColor}</td>" + "\n" + \
                    "\t\t\t\t</tr>" + "\n" + \
                    "\t\t\t\t<tr>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\"><span class=\"featureName\">Release Date: </span>{cardRDate}</td>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\"><span class=\"featureName\">Collection Date: </span>{cardCDate}</td>" + "\n" + \
                    "\t\t\t\t</tr>" + "\n" + \
                    "\t\t\t\t<tr>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\"><a href=\"PlayingCards/Backs/{cardID}.jpg\" target=\"_blank\">Image <span>({cardID}.jpg)</span></a></td>" + "\n" + \
                        "\t\t\t\t\t<td class=\"feature\">{videoFill}</td>" + "\n" + \
                    "\t\t\t\t</tr>" + "\n" + \
                "\t\t\t</table>"
```

```python
sectionFormat = "\t\t<div class=\"section\">" + "\n" + \
                    "\t\t\t<p class=\"subheader\">{cardID}</p>" + "\n" + \
                    "{contentFill}" + "\n" + \
                    "{tableFill}" + "\n" + \
                "\t\t</div>"
```

```python
htmlSections = []

for index, deck in collection.iterrows():
    cardID = deck['ID']
    cardStyle = deck['Style']
    cardPublisher = deck['Publisher']
    cardColor = deck['Color']
    cardStock = deck['Stock']
    try:
        cardRDate = deck['Release Date'].strftime("%B %Y")
    except:
        cardRDate = deck['Release Date']
    try:
        cardCDate = deck['Release Date'].strftime("%B %Y")
    except:
        cardCDate = deck['Release Date']
    
    if deck['Video']=='Yes':
        videoFill = videoLink.format(cardID=cardID)
        contentFill = videoFormat.format(cardID=cardID)
    else:
        videoFill = videoNoLink
        contentFill = imageFormat.format(cardID=cardID)
        
    tableFill = tableFormat.format(cardID=cardID, cardStyle=cardStyle, cardPublisher=cardPublisher, cardColor=cardColor, cardStock=cardStock, cardRDate=cardRDate, cardCDate=cardCDate, videoFill=videoFill)
    
    htmlSections.append(sectionFormat.format(cardID=cardID, contentFill=contentFill, tableFill=tableFill))
```

```python
htmlOutput = "<!-- Auto Generated -->\n\n"

for section in htmlSections:
    htmlOutput = htmlOutput + section + "\n\n"
    
htmlOutput = htmlOutput + "<!-- End Auto Generated -->"
```
